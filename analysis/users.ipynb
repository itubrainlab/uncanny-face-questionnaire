```python
import pandas as pd
import glob
```

```python
user_files = glob.glob("../export/*_user.csv")
user_files.sort()
user_files[-1]
```

```python
users = pd.read_csv(user_files[-1], index_col=["index"],
                    names=["index", "id", "uid", "age", "gender", "language", "timestamp"])
```

# Age Distribution

```python
users["age"].hist()
```

# Gender Distribution

```python
users["gender"].map({0: "None", 1: "Female", 2: "Male", 3: "Other"}).hist()
```

# Lnguage Distribution

```python
users["language"].map({0: "None", 1: "English", 2: "Non-english"}).hist(bins=3)
```
